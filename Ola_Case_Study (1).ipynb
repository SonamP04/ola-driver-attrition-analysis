```python
before = df[['age','gender']].isnull().sum()

#missing value strategy
df['age'] = df['age'].fillna(df['age'].median())
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])

# Target Variable
df['left'] = df['lastworkingdate'].notnull().astype(int)

print(df['left'].value_counts())
print(df[['age','gender','lastworkingdate','left']].head(10))

after = df[['age','gender']].isnull().sum()
pd.DataFrame({'before': before, 'after': after})
```

```python
# Outlier detection & treatment :

import matplotlib.pyplot as plt
import seaborn as sns

def plot_before_after(df, col):

    # copy original for comparison
    df_copy = df.copy()

    p1, p99 = df_copy[col].quantile([0.01, 0.99])

    # capped column
    capped_col = col + "_capped"
    df_copy[capped_col] = df_copy[col].clip(lower=p1, upper=p99)

    # count of capped values
    capped_count = ( (df_copy[col] < p1) | (df_copy[col] > p99) ).sum()
    print(f"{col}: {capped_count} values capped (outside 1st–99th percentile)")

    # plot boxplots
    fig, axes = plt.subplots(1, 2, figsize=(12,5))
    sns.boxplot(y=df_copy[col], ax=axes[0], color="lightcoral")
    axes[0].set_title(f"Before Capping - {col}")
    sns.boxplot(y=df_copy[capped_col], ax=axes[1], color="lightblue")
    axes[1].set_title(f"After Capping - {col}")
    plt.show()

    # plot histograms
    fig, axes = plt.subplots(1, 2, figsize=(12,5))
    sns.histplot(df_copy[col], kde=True, ax=axes[0], color="salmon")
    axes[0].set_title(f"Before Capping - {col}")
    sns.histplot(df_copy[capped_col], kde=True, ax=axes[1], color="skyblue")
    axes[1].set_title(f"After Capping - {col}")
    plt.show()

plot_before_after(df, 'total_business_value')
plot_before_after(df, 'income')

```

```python
# feature engineering
df['end_date'] = df['lastworkingdate'].fillna(df['reporting_date'])
df['tenure_days'] = (df['end_date'] - df['dateofjoining']).dt.days
df['tenure_days'] = df['tenure_days'].clip(lower=0)
df['tenure_months'] = (df['tenure_days'] / 30).round(1)

df = df.sort_values(['driver_id','reporting_date'])
df['prev_quarterly_rating'] = df.groupby('driver_id')['quarterly_rating'].shift(1)
df['rating_increased'] = np.where(df['quarterly_rating'] > df['prev_quarterly_rating'], 1, 0)
df['prev_income'] = df.groupby('driver_id')['income'].shift(1)
df['income_increased'] = np.where(df['income'] > df['prev_income'], 1, 0)

#as 1st records has no previous
df['rating_increased'] = df['rating_increased'].fillna(0).astype(int)
df['income_increased'] = df['income_increased'].fillna(0).astype(int)

#Time features
df['reporting_month'] = df['reporting_date'].dt.month
df['reporting_year'] = df['reporting_date'].dt.year

#Tenure bins
bins = [0,1,3,6,12,36,120]
labels = ['<1m','1-3m','3-6m','6-12m','12-36m','36m+']
df['tenure_bucket'] = pd.cut(df['tenure_months'], bins=bins, labels=labels, include_lowest=True)

print(df[['driver_id', 'dateofjoining', 'lastworkingdate', 'reporting_date', 'end_date', 'tenure_days', 'tenure_months']].head(10))
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Distribution of numeric features
plt.figure(figsize=(12,4))
sns.histplot(df['age'], kde=False)
plt.title('Age distribution')
plt.show()

# Churn by tenure bucket
churn_by_tenure = df.groupby('tenure_bucket')['left'].mean().reset_index()
sns.barplot(x='tenure_bucket', y='left', data=churn_by_tenure)
plt.title('Churn rate by tenure bucket')
plt.show()

# Monthly join/leave counts
joins = df.dropna(subset=['dateofjoining']).groupby(df['dateofjoining'].dt.to_period('M'))['driver_id'].nunique()
leaves = df.dropna(subset=['lastworkingdate']).groupby(df['lastworkingdate'].dt.to_period('M'))['driver_id'].nunique()

joins.plot(kind='line', label='joins'); leaves.plot(kind='line', label='leaves'); plt.legend(); plt.show()
```

```python
#Predictive Analysis

# Train/test split by Driver_ID to avoid leakage.

from sklearn.model_selection import GroupShuffleSplit
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
groups = df['driver_id']
train_idx, test_idx = next(gss.split(df, groups=groups))
train = df.iloc[train_idx]; test = df.iloc[test_idx]


```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc

features = [f for f in ['age','income','grade','tenure_months','quarterly_rating',
                        'tenure_rating_interaction','tbv_3m_avg','recent_income_drop','recent_rating_drop'] if f in df.columns]
X_train = train[features].fillna(0); y_train = train['left']
X_test = test[features].fillna(0); y_test = test['left']

# Baseline logistic
lr = LogisticRegression(max_iter=1000, class_weight='balanced')
lr.fit(X_train, y_train)
proba_lr = lr.predict_proba(X_test)[:,1]

# RF
rf = RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)
proba_rf = rf.predict_proba(X_test)[:,1]

# XGBoost
xgb_clf = xgb.XGBClassifier(n_estimators=200, use_label_encoder=False, eval_metric='auc', scale_pos_weight=(y_train==0).sum()/(y_train==1).sum())
xgb_clf.fit(X_train, y_train)
proba_xgb = xgb_clf.predict_proba(X_test)[:,1]

def pr_auc(y_true, y_scores):
    p, r, _ = precision_recall_curve(y_true, y_scores)
    return auc(r, p)

for name, proba in [('LR',proba_lr),('RF',proba_rf),('XGB',proba_xgb)]:
    print(name, "ROC AUC:", round(roc_auc_score(y_test, proba),3), "PR AUC:", round(pr_auc(y_test,proba),3))
```

```python
#top features and a short interpretation

# feature importance (RF)
import pandas as pd
feat_importance = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)
print(feat_importance.head(15))

# SHAP (for XGBoost — faster)
import shap
explainer = shap.TreeExplainer(xgb_clf)
X_test_small = X_test.sample(min(500, len(X_test)))
shap_values = explainer.shap_values(X_test_small)
shap.summary_plot(shap_values, X_test_small, feature_names=features, plot_type="bar")
```

**Insights & Recommendations**

The churn rate analysis by tenure shows that most drivers leave within their first 6 months, with the peak churn occurring between 3–6 months (~19%). After 1 year, attrition declines sharply, indicating that retaining drivers in the early months is critical for long-term workforce stability.

The company experienced rapid growth in driver sign-ups from 2017 onward, attrition also rose sharply from 2018, with the number of drivers leaving often close to or exceeding new joins. This indicates that despite strong hiring, retention challenges threaten long-term stability.

The analysis shows negligible difference in business value across education levels, indicating that education is not a key driver of performance. However, significant variation exists across cities — with some cities (e.g., C29, C12) yielding higher median business value. This highlights that geographic location is a more critical factor to consider for driver management and retention strategies.

There is a weak positive correlation between age and income (Pearson r ≈ 0.19, Spearman ρ ≈ 0.20). Older drivers tend to earn slightly more than younger drivers, but the relationship is not strong, suggesting that other factors (such as city, tenure, or performance ratings) play a bigger role in determining income.

Drivers with quarterly rating = 1 have a substantially higher risk of leaving (≈18.6% churn) and much shorter tenure (≈10.5 months) compared to higher-rated peers.I recommend a focused, low-cost pilot targeting rating=1 drivers in their early months (coaching + light incentives).Rating should be used as a primary signal in operational risk-scoring to prioritize retention efforts.

The logistic regression model achieved a good ability to distinguish between drivers who leave and those who stay (ROC AUC = 0.85). It successfully identified nearly 88% of actual leavers, though with low precision (19%), meaning many false positives. This suggests the model is effective at flagging potential churn risk but should not be used as a direct predictor. Instead, Ola can apply the model as a risk-scoring tool to focus retention efforts on the top 5–10% highest-risk drivers. Interventions such as low-cost coaching, SMS nudges, or performance incentives can then be piloted.

We propose testing targeted interventions (driver coaching + ₹X incentive) through an A/B experiment on the top 10% high-risk drivers. By comparing retention rates of Treatment vs Control over 8–12 weeks, Ola can validate if churn reduction is significant and cost-effective before scaling the program.
